```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get('GEMINI_API'))
```

```python
def generate_email(email_type,context,sender_name,recipient_name,tone="professional",additional_details=""):

    """
    Generate a professional email.

    Args:
        email_type        : e.g. 'cold outreach', 'follow-up', 'apology',
                            'job application', 'sales pitch', 'newsletter',
                            'complaint', 'thank you'
        context           : What the email is about / the main purpose
        sender_name       : Name of the person sending the email
        recipient_name    : Name of the recipient (or 'Hiring Manager', 'Team', etc.)
        tone              : 'professional', 'friendly', 'formal', 'persuasive', 'empathetic'
        additional_details: Any extra info to include (product name, date, offer, etc.)

    Returns:
        str: A complete email (subject line + body)
    """

    system_prompt = (
        "You are an expert email copywriter who crafts compelling, clear, and effective emails. "
        "Every email you write includes: a strong subject line, a proper greeting, "
        "a well-structured body, and a professional sign-off. "
        "You tailor the language and tone precisely to the email type and audience. "
        "You never use filler words like 'I hope this email finds you well'."
    )

    user_prompt = f"""
    Write a complete email with the following details:

    Email Type         : {email_type}
    Purpose / Context  : {context}
    Sender Name        : {sender_name}
    Recipient Name     : {recipient_name}
    Tone               : {tone}
    Additional Details : {additional_details if additional_details else "None"}

    Format the output as:
    Subject: <subject line>

    <email body>
    """

    model = genai.GenerativeModel(
        model_name='gemini-2.5-flash',
        system_instruction=system_prompt
    )

    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.8,
            "max_output_tokens": 600,
        }
    )

    return response.text
```

```python
if __name__ == "__main__":
    email = generate_email(
        email_type="cold outreach",
        context="Pitching an AI-powered productivity tool to a startup founder",
        sender_name="Vidit Shah",
        recipient_name="Tapan Mehta",
        tone="persuasive and friendly",
        additional_details="The tool is called FlowAI, it integrates with Notion and Slack, free trial available",
    )
    print("=" * 60)
    print(email)
    print("=" * 60)

    print("\n── Interactive Email Generator ──")
    email_types = [
        "cold outreach", "follow-up", "apology", "job application",
        "sales pitch", "newsletter", "complaint", "thank you", "meeting request"
    ]
    print("Email types:", ", ".join(email_types))

    etype   = input("\nEmail type       : ")
    ctx     = input("Purpose/context  : ")
    sender  = input("Your name        : ")
    recip   = input("Recipient name   : ")
    tone    = input("Tone (professional / friendly / formal / persuasive): ") or "professional"
    extras  = input("Additional details (optional): ")

    print("\nGenerating your email...\n")
    print(generate_email(etype, ctx, sender, recip, tone, extras))
```
