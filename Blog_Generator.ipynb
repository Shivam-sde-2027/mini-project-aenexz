```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get('GEMINI_API'))
```

```python
def generate_blog(topic, tone, target_audience, word_count=500):
    """
    Generate a blog post.

    Args:
        topic           : What the blog is about
        tone            : e.g. 'professional', 'casual', 'humorous', 'informative'
        target_audience : e.g. 'beginners', 'developers', 'entrepreneurs'
        word_count      : Approximate length (default 500 words)

    Returns:
        str: Generated blog post
    """
    system_prompt = (
        "You are an expert blog writer with years of experience in content marketing. "
        "You write engaging, SEO-friendly blog posts with clear structure: "
        "a catchy title, an introduction, well-organized body sections with subheadings, "
        "and a strong conclusion with a call to action. "
        "Your writing is original, informative, and tailored to the target audience."
    )

    user_prompt = f"""
    Write a complete blog post with the following details:

    Topic           : {topic}
    Tone            : {tone}
    Target Audience : {target_audience}
    Approximate Length: {word_count} words

    Structure the blog with:
    - A compelling title
    - An engaging introduction
    - 3-4 body sections, each with a subheading
    - A conclusion with a call to action
    """

    model = genai.GenerativeModel(
        model_name='gemini-2.5-flash-lite',
        system_instruction=system_prompt
    )

    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.8,
            "max_output_tokens": 1200,
        }
    )

    return response.text
```

```python
if __name__ == "__main__":
    blog = generate_blog(
        topic="Why Python is the best language for beginners",
        tone="casual and friendly",
        target_audience="college students new to programming",
        word_count=400,
    )
    print("=" * 60)
    print(blog)
    print("=" * 60)

    print("\n── Interactive Blog Generator ──")
    topic    = input("Blog topic       : ")
    tone     = input("Tone (e.g. professional / casual / humorous): ")
    audience = input("Target audience  : ")
    length   = input("Word count (default 500): ") or "500"

    print("\nGenerating your blog post...\n")
    print(generate_blog(topic, tone, audience, int(length)))
```
